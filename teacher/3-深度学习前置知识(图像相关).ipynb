## numpy常见随机函数

```python
# 下面介绍生成随机整数数组以及相关操作

# 首先使用np.randint函数生成一些随机整数
# np.random.randint函数用于生成指定范围和数量的随机整数数组，这里的参数分别表示最小值（包含）、最大值（不包含）以及生成的随机整数数量
arr = np.random.randint(0, 50, 10)

# 打印一下生成的随机整数数组，查看具体生成的随机数值情况
print(arr)
```

```python
# 使用max获取最大值，这可以快速找到数组中的最大数值，在分析医学数据中的极值情况等场景会用到
print(arr.max())

# 再使用argmax() 获取最大值的索引，知道最大值在数组中的位置有时候对进一步分析数据、定位特定数据点很有帮助
print(arr.argmax())

# 使用min函数获取最小值，与获取最大值相对应，能帮助我们了解数据的范围边界情况
print(arr.min())

# 使用argmin获取最小值索引，确定最小值所在位置
print(arr.argmin())

# 使用mean()方法获取取平均值，平均值是统计分析中常用的指标，比如分析医学检测指标的平均水平等情况
print(arr.mean())
```

## shape操作

```python
# 如果要获取numpy数组的大小，使用numpy.shape，它返回的是一个元组，表示数组各维度的大小，对于一维数组就是元素个数，二维数组就是（行数，列数）等形式
print(arr.shape)

# 也可以使用reshape函数转换数组的形状，这在调整数据维度以适应不同算法输入要求等场景非常有用，比如我将arr转换成5*2的数组
print(arr.reshape((5, 2)))

# 可以看到变成了5行2列，改变形状后的数组元素顺序是按照特定规则重新排列的，遵循C语言风格的行优先顺序（在内存中连续存储行元素）
print(arr.reshape((2, 5)))

# 尝试错误变形2*10看看效果，当我们尝试将数组重塑为不符合元素数量匹配的形状时，会抛出ValueError异常，这里通过try-except语句捕获并打印异常信息，让我们清楚知道哪里出了问题
try:
    print(arr.reshape((2, 10)))
except ValueError as e:
    print(e)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# 获取图像的高度和宽度
height, width, _ = img_arr.shape

# 定义特定区域（这里是中间部分）
start_x = int(width / 4)
end_x = int(width * 3 / 4)
start_y = int(height / 4)
end_y = int(height * 3 / 4)

# 提取特定区域
specific_region = img_arr[start_y:end_y, start_x:end_x, :]

# 计算特定区域的平均颜色值（这里是在三个颜色通道上分别求平均）
average_color = np.mean(specific_region, axis=(0, 1))

# 打印平均颜色值
print("Average color in the specific region:", average_color)

# 使用matplotlib绘制图像，并标记出特定区域
plt.imshow(img_arr)
rect = plt.Rectangle((start_x, start_y), end_x - start_x, end_y - start_y,
                     linewidth=2, edgecolor='r', facecolor='none')
plt.gca().add_patch(rect)
plt.show()
```

```python

# 我们继续对这个图片操作，先使用numpy的copy方法复制一份原图，复制操作很重要，它能让我们在原始图像数据基础上进行各种尝试性操作，
# 而不会影响到原始图像，方便对比操作前后的变化以及进行多次不同的实验
img_arr_copy = img_arr.copy()
```

# tensor入门

## 开始之前，先讲一下numpy和tensor的关系

```python
# NumPy 和 PyTorch 都是用于科学计算的库，但它们分别在不同的领域中更为常见。
# NumPy 是 Python 的一个基础科学计算库，而 PyTorch 是一个主要用于深度学习的库。
# 尽管它们在某些方面有重叠，但它们之间存在一些关键的区别。
```

```python
# 示例 1：创建数组/张量
import numpy as np
import torch

# 创建一个 NumPy 数组
numpy_array = np.array([1, 2, 3, 4, 5])

# 创建一个 PyTorch 张量
torch_tensor = torch.tensor([1, 2, 3, 4, 5])

print("NumPy Array:", numpy_array, " type = ",type(numpy_array))
print("PyTorch Tensor:", torch_tensor, " type = ",type(torch_tensor))
```

```python
# NumPy 数组运算
numpy_result = np.array([1, 2, 3]) + np.array([4, 5, 6])
print("NumPy Result:", numpy_result)

# PyTorch 张量运算
torch_result = torch.tensor([1, 2, 3]) + torch.tensor([4, 5, 6])
print("PyTorch Result:", torch_result)
```

```python
# umPy 数组和 PyTorch 张量可以互相转换。这种转换在处理数据时非常有用，
# 尤其是当你需要在 NumPy 的强大数值计算能力和 PyTorch 的自动微分及GPU加速之间切换时。
import numpy as np
import torch

# 创建一个 NumPy 数组
numpy_array = np.array([1, 2, 3, 4, 5])

# 将 NumPy 数组转换为 PyTorch 张量
torch_tensor_from_numpy = torch.from_numpy(numpy_array)

print("NumPy Array:", numpy_array)
print("PyTorch Tensor from NumPy:", torch_tensor_from_numpy)
```

```python
# 从 PyTorch 张量转换到 NumPy 数组
# 创建一个 PyTorch 张量
torch_tensor = torch.tensor([1, 2, 3, 4, 5])

# 将 PyTorch 张量转换为 NumPy 数组
numpy_array_from_torch = torch_tensor.numpy()

print("PyTorch Tensor:", torch_tensor)
print("NumPy Array from PyTorch:", numpy_array_from_torch, " type = ",type(numpy_array_from_torch))
```

```python
# 注意事项
# 1.内存共享：当从 PyTorch 张量转换到 NumPy 数组时，这两个对象会共享相同的内存空间。
# 因此，对其中一个对象的修改会反映在另一个对象上。
# 
# 2.设备兼容性：如果 PyTorch 张量位于 GPU 上，你需要先将其移动到 CPU 上，
# 然后才能转换为 NumPy 数组，因为 NumPy 不支持 GPU 操作。
```

```python
import numpy as np
# 中文比较麻烦

# 导入PIL对应包
from PIL import Image, ImageDraw, ImageFont

# 定义一个函数
def cv2AddChineseText(img, text, position, textColor=(0, 255, 0), textSize=30):
    img = Image.fromarray(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    print(type(img))
    draw = ImageDraw.Draw(img)
    # 字体的格式
    fontStyle = ImageFont.truetype(
        "./font/simsun.ttc", textSize, encoding="utf-8")
    # 绘制文本
    draw.text(position, text, textColor, font=fontStyle)
    # 转换回OpenCV格式
    return cv2.cvtColor(np.asarray(img), cv2.COLOR_RGB2BGR)
```

```python
img = cv2.imread("./histo1.png")
img = cv2AddChineseText(img, '医学 人工智能', (20,50), textColor=(255, 0, 0), textSize=30)
img_med_ai = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_med_ai)
```

```python
import cv2
import xml.etree.ElementTree as ET
import matplotlib.pyplot as plt
import os

# 读取XML文件
tree = ET.parse('BloodImage_00000_jpg.rf.67c4a4312251eaa52b6ea2f2edf4855b.xml')
root = tree.getroot()

# 读取图片路径
image_path = root.find('path').text

# 检查图片路径是否存在
if not os.path.exists(image_path):
    print(f"Error: Image path does not exist: {image_path}")
else:
    # 读取图片
    image = cv2.imread(image_path)

    # 检查图片是否成功读取
    if image is None:
        print("Error: Failed to read the image.")
    else:
        # 转换颜色格式从BGR到RGB
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # 遍历每个对象并绘制边界框
        for obj in root.findall('object'):
            name = obj.find('name').text
            bndbox = obj.find('bndbox')
            xmin = int(bndbox.find('xmin').text)
            ymin = int(bndbox.find('ymin').text)
            xmax = int(bndbox.find('xmax').text)
            ymax = int(bndbox.find('ymax').text)
            
            # 绘制矩形框
            cv2.rectangle(image, (xmin, ymin), (xmax, ymax), (0, 255, 0), 2)
            
            # 在框上方显示标签
            cv2.putText(image, name, (xmin, ymin - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        # 使用plt显示图片
        plt.imshow(image)
        plt.axis('off')  # 不显示坐标轴
        plt.show()
```

```python
# 数据集介绍
# Lung-PET-CT-Dx是一个公开肺癌目标检测数据集，该公共数据集源自 The Cancer Imaging Archive（TCIA）。
# TCIA是肿瘤研究方面的著名的医学影像公开数据库。大部分数据以 DICOM 格式存储，所有数据都经过去隐私处理。

# 该数据集由肺癌受试者的 CT 和 PET-CT DICOM 图像组成，带有 XML 注释文件（xml文件中用边界框标识了肿瘤位置和分类）。
# 这些PET/CT影像是从肺癌患者中回顾性获得的，患者均经肺活检确诊 。 根据组织病理学诊断对受试者进行分组，分为4类，“A”诊断为腺癌，
# “B”为小细胞癌，“E”为大细胞癌，“G”为鳞状细胞癌。
```
