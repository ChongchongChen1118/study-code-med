## if分支语句

```python
# if语句 分支 ： 我们是根据条件的真假来决定是否执行某段代码
# 需求: score 考试成绩
# 当成绩大于等于90分，我们认为成绩优秀
# 当成绩80-89分，我们认为成绩良好
# 当成绩大于60-79分，我们认为及格
# 成绩小于60，就是不及格
```

```python
score = 20
```

```python
score >= 90
```

```python
score >= 80
```

```python
if score >= 90:
    print("优秀")
    print("再接再厉，更上一层楼")
elif score >= 80:
    print("成绩良好")
elif score >= 60:
    print("及格")
else:
    print("不及格")
```

## 循环语句


```python
# 循环：重复执行某种操作,根据条件，重复执行某种操作
# for循环
# while循环
```

```python
# 对象  男朋友/女朋友
# 不小心，某一天，惹对象生气
# 对象说：你要道歉100句，我就原谅你
```

```python
# for 循环来实现
# 1,2,3,4,....100
for i in range(1,101):
    print("对不起",i,"次")
```

```python
# for循环可以遍历序列(如列表、元组、字典、字符串)或者迭代器
fruits = ['apple','banana','orange']
```

```python
fruits
```

```python
type(fruits) # list 列表
```

```python
# for循环我们来遍历这个fruits列表
for fr in fruits:
    print(fr)
```

```python
# while循环
# while循环会持续执行,直到条件不再成立 False
i = 1
while i < 5:
    print(i)
    i = i + 1
```

```python
## for while循环，我们是可以搭配上if语句一起使用的
for i in range(1,11):
    if i % 2 == 0:
        print(i)
```

```python
## for while我们是可以提前终止的
# break可以退出整个循环
for i in range(1,11):
    if i == 3:
        break;
    print(i)
    
```

```python
# continue 中止某次循环,继续下一次循环
for i in range(1,11):
    if i == 3:
        continue;
    print(i)
    
```

```python
## 带有可变参数的函数
# 新手 print()
# debug -> 调式
# *args
def sum_numbers(*ars):
    print("ars = ",ars)
    total = 0
    for number in ars:
        total = total + number
    return total
```

```python
sum_numbers(1,2,3,4,5,6,7,8)
```

```python
## 带有关键字参数的函数
def print_person(name, age):
    print(f"Name: {name}, Age: {age}")
```

```python
print_person(name='桑尼', age=38)
```

```python
print_person(age=38,name="桑尼")
```

```python
## 带有可变关键字参数的函数
def print_info(**kwargs):
    print("kwargs = ",kwargs)
    for key,value in kwargs.items():
        print(f"{key}: {value}")
```

```python
print_info(name="汤姆", age=36, job="律师", hobby='看书')
```

```python
## 面向对象的思想的三大核心
# 继承、封装、多态
```

```python
# 继承：子承父业
# 你可以继承你父亲的东西
# 我们其实在代码里面允许，一个类(子类)去继承另外一个类（父类）的属性和方法
```

```python
# 定义一个
class Dog:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def bark(self):
        print("狗汪汪汪...")
```

```python
# 我定义了一个Husky类并且这个类是继承Dog
# super() 父类  Dog
class Husky(Dog):
    def __init__(self,name,age,furry_color):
#         super().__init__(name,age)
        self.name = name
        self.age = age
        self.furry_color = furry_color
    def run(self):
        print("哈士奇在跑步...")
```

```python
my_husky = Husky("来福",8,'棕黄色')
```

```python
my_husky.name
```

```python
my_husky.age
```

```python
my_husky.furry_color
```

```python
my_husky.bark()
```

```python
# 封装
# 封装：其实就是把数据（属性）和方法（行为）打包在一起，通过封装我们其实可以实现对属性的访问控制
```

```python
# 账户类
# 账户属于谁的: owner
# balance： 多少余额
# 把balance封装成私有属性,外界不能直接通过对象.__balance直接获取
# __balance
class Account:
    def __init__(self,owner, balance=0):
        self.owner = owner
        self.__balance = balance
    def get_balance(self):
        return self.__balance
```

```python
my_account = Account("麦克",1000)
```

```python
my_account.owner
```

```python
my_account.get_balance()
```

```python
## 多态
# 多态：允许你使用统一的接口去访问不同类的对象
```

```python
# 动物类
class Animal:
    def speak(self):
        print("动物叫....")
```

```python
class Dog(Animal):
    def speak(self):
        print("汪汪汪...")
```

```python
class Cat(Animal):
    def speak(self):
        print("喵喵喵......")
```

```python
my_dog = Dog()
```

```python
my_dog.speak()
```

```python
my_cat = Cat()
```

```python
my_cat.speak()
```

```python
def animal_speak(animal):
    animal.speak()
```

```python
animal_speak(my_dog)
```

```python
animal_speak(my_cat)
```
